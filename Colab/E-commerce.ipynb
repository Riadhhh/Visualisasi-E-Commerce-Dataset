### 1.2 Menilai Data

1.2.1 Memeriksa Tipe Data

```python
for df_name, df in data.items():
  print('\n', data[df_name].info())
```

```python
datetime_oo = ["order_purchase_timestamp","order_approved_at","order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"]
for column in datetime_oo:
  data['orders'][column] = pd.to_datetime(data['orders'][column])

datetime_oi = ["shipping_limit_date"]

for column in datetime_oi:
  data['items'][column] = pd.to_datetime(data['items'][column])

datetime_or = ["review_creation_date","review_answer_timestamp"]

for column in datetime_or:
  data['reviews'][column] = pd.to_datetime(data['reviews'][column])
```

1.2.2 Memeriksa Duplikasi

```python
# Menginisialisasi dictionary untuk menyimpan jumlah duplikat dan total jumlah sampel. Dictionary tu cmn bos? kamus kah?
duplicate_counts = {}
total_sample_counts = {}

# Menghitung dan menyimpan jumlah duplikat and total jumlah sampel untuk setiap DataFrame. Apalagi DataFrame tuch bolo?
for df_name, df in data.items():
    duplicate_counts[df_name] = df.duplicated().sum()
    total_sample_counts[df_name] = len(df)  # Calculate the total number of samples

# Membuat DataFrame/Tabel apalatu dari Dictionaries
duplicate_counts_df = pd.DataFrame.from_dict(duplicate_counts, orient='index', columns=['Duplicate Count'])
total_sample_counts_df = pd.DataFrame.from_dict(total_sample_counts, orient='index', columns=['Total Sample Count'])

# Gabungkan Dua DataFrames secara Horizontal bolo
pd.concat([total_sample_counts_df, duplicate_counts_df], axis=1)
```

1.2.3 Memeriksa Nilai yang kosong

```python
# Menginisialisasi Dictionary untuk menyimpan jumlah nilai kosong untuk setiap DataFrame
null_counts = {}

# Menghitung dan menyimpan jumlah nilai kosong untuk setiap DataFrame
for df_name, df in data.items():
    null_counts[df_name] = df.isna().sum()
    print(df_name, '\n', null_counts[df_name], '\n')
```

### **1.3 Data Cleaning**

1.3.1 Urutan



*    Ni kumpulan data yang paling penting karena seperti tabel penghubung, dan terhubung ke banyak kumpulan data lain melalui order_id.



```python
# untuk analisis ni, kupilih pesanan yg udah diantar (ke hatimu) aja
print(data['orders']['order_status'].value_counts())
```

```python
print('Meskipun kita hanya mengambil pesanan yang berstatus (telah dikirim), kita masih memiliki ' +
      str(round((data['orders'].order_status.value_counts()['delivered']/len(data['orders']))*100,2)) +
      '% dari keseluruhan data.')
```

```python
# jadi, kuambil dlu semua order_id yang statusnya delivered (dikirim)
delivered = data['orders']['order_id'][data['orders']['order_status'] != 'delivered'].values

# skrg, filter setiap dataframe yang memiliki order_id sebagai foreign key (dari setiap dataset bosku)
for key, value in data.items():
    if 'order_id' in value.columns:
        value.drop(value.loc[value['order_id'].isin(delivered)].index, inplace = True)
```

```python
# memeriksa nan (Not A Number) di setiap Dataset (gk tau NaN? NaN tu singkatnya data yang hilang bolo)
for key, value in data.items():
    for col in value.columns:
        # cetak kolom yang datanya hilang aja (dalam persen)
        if value[col].isnull().sum():
            print(col + ' (%): ' + str((sum(value[col].isnull())/len(value))*100))
```

```python
# klo gtu, kita urus dlu orders yang ilang. kita bisa lihat nans di list order tdi tu gk bertumpukan dan yang kuhiraukan cuma delivered_customer_date nya aja, yg lain los
print('Cuma ada ' + str(data['orders'].order_delivered_customer_date.isnull().sum()) +
      ' pesanan yang kehilangan tanggal.')
```

```python
# sama kek sebelumnya, kita hapus dulu yg lain, pertama-tama kita ambil order_id yang statusnya "delivered" (telah dikirim)
null_deliveries = data['orders']['order_id'][data['orders']['order_delivered_customer_date'].isnull()].values

# kek kublg tdi sm kek sblmnya filter dataframe lain yang memiliki order_id sebagai foreign key
for key, value in data.items():
    if 'order_id' in value.columns:
        value.drop(value.loc[value['order_id'].isin(null_deliveries)].index, inplace = True)
```

```python
# disini kita hanya tertuju pada delivered on time, bukan tanggal deliverynya (pemesanan)
data['orders']['delivered_on_time'] = np.where(data['orders']['order_delivered_customer_date'] < data['orders']['order_estimated_delivery_date'], 'On Time', 'Late')
print(data['orders']['delivered_on_time'].value_counts())
```

1.3.2 Reviews

```python
print('Review cols: ' , data['reviews'].columns.values)

# 88,36 dan sekian% dari review_comment_title tu NaN ato kosong, jadi..ya hapus aja la, nnti takutnya resiko mengganggu analisis karena banyak yg hilang
review_drop_cols = ['review_comment_title']
data['reviews'].drop(review_drop_cols, axis= 1, inplace= True)
```

```python
# jad1 di review_comment_message hanya kurang dari separuh yang mengisi ulasan, jdi kita set aja jadi 0 mate
data['reviews']['review_comment_message'] = np.where(data['reviews']['review_comment_message'].isnull(), 0, 1)
```

1.3.3 Products

```python
print('Product cols: ' , data['products'].columns.values)

# Seperti yang kita lihat, kolom product banyak kali nans nya, jadi gak tau apa yang mau dibuat sm mereka kecuali dihapus aja, jdi ni mau ku drop la ya bro
product_drop_cols = ['product_name_lenght',
                     'product_description_lenght',
                     'product_weight_g',
                     'product_length_cm',
                     'product_height_cm',
                     'product_width_cm']

data['products'].drop(product_drop_cols, axis= 1, inplace= True)
```

```python
# jadi abis ngecek kolom kategori produk, NaN yang ada disitu mungkin cuma gk tercatat, jadi kubuat aja jadi "other" (karena ya apalagi kategorinya brok)
data['products']['product_category_name'].fillna(value='other', inplace=True)
print(data['products']['product_category_name'].isna().sum())
```

```python
# jadi gdk NaN di kolom jumlah foto produk, agak membangongkan. jadi pasti ada aja produk yang gdk gambar/fotonya, jdi langsung ajala buat jadi 0 yang kosong
data['products']['product_photos_qty'].fillna(value=0, inplace=True)
print(data['products']['product_photos_qty'].isna().sum())
```
